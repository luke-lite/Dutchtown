```python
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', None)
```

```python
stl_df = pd.read_csv('stl_crime_data_2021-2023.csv', index_col=0)
gj_df = pd.read_csv('gravois-jefferson_crime_data_2021-2023.csv', index_col=0)
```

```python
gj_df
```

```python
# gj_df.index = pd.to_datetime(gj_df.index)

gj_df['Person'].pct_change(periods=12)
```

```python
pct_chg_df = pd.DataFrame(columns=['PctChg_Person', 'PctChg_Property', 'PctChg_Society', 'PctChg_Unspecified', 'PctChg_Total'])
pct_chg_df['PctChg_Person'] = gj_df['Person'].pct_change(periods=12)
pct_chg_df['PctChg_Property'] = gj_df['Property'].pct_change(periods=12)
pct_chg_df['PctChg_Society'] = gj_df['Society'].pct_change(periods=12)
pct_chg_df['PctChg_Unspecified'] = gj_df['Unspecified'].pct_change(periods=12)
pct_chg_df['PctChg_Total'] = gj_df['Total'].pct_change(periods=12)
pct_chg_df.to_csv('gj_pct_chg_df.csv')
```
