```python
nibrs_df = pd.read_csv('crime_data/nibrs_thru_dec_2023.csv')
nibrs_df
```

```python
# drop neighborhood 'G-88', it is outside city limit (126 rows)
nibrs_df = nibrs_df[nibrs_df['neighborhood'] != 'G-88']
# drop nan value neighborhoods (2 rows)
nibrs_df.dropna(subset=['neighborhood'], inplace=True)
nibrs_df['neighborhood'] = nibrs_df['neighborhood'].astype('int64')
```

```python
nibrs_df
```

```python
nibrs_df = nibrs_df.merge(stl_nbhd_population, how='left', left_on='neighborhood', right_on='NBHD_Code')
nibrs_df
```

```python
nibrs_df.drop('NBHD_Code', axis=1, inplace=True)
nibrs_df.rename(columns={'neighborhood':'nbhd_code', 'Neighborhood':'nbhd', 'Population': 'population'}, inplace=True)
nibrs_df
```

```python
nibrs_codes_dict = {'Person': ['09A', '09B', '09C', '09D', '100', '11A', '11B', '11C', '11D', '13A', '13A*', '13B', '13C', '36A', '36B', '64A', '64B'], 
                    'Property': ['120', '120*', '200', '210', '220', '23A', '23B', '23C', '23D', '23E', '23F', '23G', '23H', '240', '250', '26A', '26B', '26C', '26D', '26E', '26F', '26G', '270', '280', '290', '510', '90A'],
                    'Society': ['35A', '35B', '370', '39A', '39B', '39C', '39D', '40A', '40B', '40C', '520', '720', '90B', '90C', '90D', '90E', '90F', '90G', '90H', '90J'],
                    'Unspecified': ['90Z']}
violent_crimes = ['09A','11A','13A','13A*','120','120*']
```

```python
def get_key(val):
    for key, value in nibrs_codes_dict.items():
        for i in value:
            if val == i:
                return key

get_key('90Z')
```

```python
# nibrs_df['nibrs_cat'] = 
get_key(nibrs_df['nibrs_code'][0])
```

```python
nibrs_cat_list = []
for i in nibrs_df['nibrs_code']:
    cat = get_key(i)
    nibrs_cat_list.append(cat)
nibrs_cat_list
```

```python
nibrs_df['is_violent'] = nibrs_df['nibrs_code'].isin(violent_crimes)
```

```python
nibrs_df['nibrs_cat'] = nibrs_cat_list
nibrs_df
```

```python
nibrs_df.to_csv('stl_crime_raw_tableau.csv')
```

```python
stl_crime_counts = pd.read_csv('stl_crime_data_2021-2023.csv')
stl_crime_counts
```

```python
stl_crime_counts['month'] = pd.to_datetime(stl_crime_counts['month'])
stl_crime_counts
```

```python
stl_crime_counts.set_index('month', inplace=True)
stl_crime_counts.groupby(pd.Grouper(freq='M'))
```

```python
stl_crime_counts['Total_YTD'] = stl_crime_counts.groupby(pd.Grouper(freq='Y'))['Total'].cumsum()
stl_crime_counts['Violent_YTD'] = stl_crime_counts.groupby(pd.Grouper(freq='Y'))['Violent'].cumsum()
stl_crime_counts['Unspecified_YTD'] = stl_crime_counts.groupby(pd.Grouper(freq='Y'))['Unspecified'].cumsum()
stl_crime_counts['Society_YTD'] = stl_crime_counts.groupby(pd.Grouper(freq='Y'))['Society'].cumsum()
stl_crime_counts['Property_YTD'] = stl_crime_counts.groupby(pd.Grouper(freq='Y'))['Property'].cumsum()
stl_crime_counts['Person_YTD'] = stl_crime_counts.groupby(pd.Grouper(freq='Y'))['Person'].cumsum()
stl_crime_counts
```

```python
stl_crime_counts.to_csv('stl_crime_count_tableau.csv')
```

```python
shifted_stl_crime_counts = stl_crime_counts.shift(12)
shifted_stl_crime_counts
```

```python
crime_cats = ['Person', 'Property', 'Society', 'Unspecified', 'Violent', 'Total']

prct_change_df = round(100*(1 - shifted_stl_crime_counts[crime_cats] / stl_crime_counts[crime_cats]), 1)
prct_change_df
```
